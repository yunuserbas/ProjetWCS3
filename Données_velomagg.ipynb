```python
import requests
import json
import pandas as pd

url = "https://www.herault-data.fr/api/records/1.0/search/?dataset=tam_mmm_velomag&q=&lang=fr&rows=56&facet=name&facet=commune&refine.commune=Montpellier"
page = requests.get(url)
data = page.json()
```

```python
df = pd.json_normalize(data['records']).drop(['datasetid','recordid'], axis=1)

df.rename(columns={'fields.pourcentage_de_velos_restants': 'pourcentage_de_velos_restants', 'fields.num_docks_available': 'num_docks_available', 'fields.coord': 'coord',
                   'fields.commune': 'commune', 'fields.capacity': 'capacity', 'fields.num_bikes_available': 'num_bikes_available', 'fields.geopoint': 'geopoint',
                   'fields.name': 'adresse', 'fields.num_bikes_disabled': 'num_bikes_disabled', 'fields.num_bikes_available': 'num_bikes_available'}, inplace=True)
df
```

```python
from sqlalchemy import create_engine

import pymysql

import pandas as pd

import schedule
import time


def  load_data():
    link ="https://www.herault-data.fr/api/records/1.0/search/?dataset=tam_mmm_velomag&q=&lang=fr&rows=56&facet=name&facet=commune&refine.commune=Montpellier"
    # This function gets the file content from the URL
    r = requests.get(link)
    # You can access to the text content with this function
    data=json.loads( r.text)
    pd.json_normalize(r.json())
    df=pd.json_normalize(data["records"],meta='fields')
    df = pd.json_normalize(data['records']).drop(['datasetid','recordid'], axis=1)
    df.rename(columns={'fields.pourcentage_de_velos_restants': 'pourcentage_de_velos_restants', 'fields.num_docks_available': 'num_docks_available', 'fields.coord': 'coord',
                   'fields.commune': 'commune', 'fields.capacity': 'capacity', 'fields.num_bikes_available': 'num_bikes_available', 'fields.geopoint': 'geopoint',
                   'fields.name': 'adresse', 'fields.num_bikes_disabled': 'num_bikes_disabled', 'fields.num_bikes_available': 'num_bikes_available'}, inplace=True)
    df["latitude"] = df["geopoint"].apply(lat)
    df["longitude"] = df["geopoint"].apply(lon)
    df['record_timestamp'] = df['record_timestamp'].apply(date)
    data_velo = df[[ 'adresse', 'num_docks_available', 'capacity', 'num_bikes_available', 'num_bikes_disabled', 'pourcentage_de_velos_restants', 'latitude', 'longitude', 'record_timestamp']]
    
    
    tableName   = "data_velo"
    dataFrame   =data_velo
    sqlEngine       = create_engine('mysql+pymysql://user:password@127.0.0.1/projetapi?charset=utf8mb4', max_overflow=20)
    dbConnection    = sqlEngine.connect()
    try:
                       frame           = dataFrame.to_sql(tableName, dbConnection, if_exists='replace');
    except ValueError as vx:
                       print(vx)
    except Exception as ex:
                       print(ex)
    else:
                       print("Table %s)created successfully."%tableName);
    finally:
                       dbConnection.close()

```
