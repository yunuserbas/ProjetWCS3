```python
from sqlalchemy import create_engine

import pymysql

import pandas as pd

import schedule
import time


def  load_data():
    link='https://www.herault-data.fr/api/records/1.0/search/?dataset=parkingtpsreel_modele0&q=&rows=16&facet=name&facet=commune&facet=domanialit&facet=proprietai&refine.commune=Montpellier&refine.domanialit=Public'
    r = requests.get(link)
    # You can access to the text content with this function
    data = json.loads( r.text)
    # And you can import JSON object to pandas
    pd.json_normalize(r.json())
    df_parking=pd.json_normalize(data["records"],meta='fields')
    df_parking[['lat','lon']]= df_parking['geometry.coordinates'].to_list()
    df_parking[['lat','lon']]=df_parking[['lat','lon']].astype(float)
    parking=df_parking[['record_timestamp','fields.total','fields.proprietai','fields.free','fields.adresse','fields.pourcentage_de_places_disponibles','fields.name','fields.dateinfo','lat','lon']]
    parking['record_timestamp']=parking.loc[:,'record_timestamp'].apply(lambda x: x.split('.')[0].replace('T',' '))
    
    tableName   = "parking"
    dataFrame   = parking
    dataframe   = pd.DataFrame(data = parking)
    sqlEngine       = create_engine('mysql+pymysql://name:password@127.0.0.1/projetapi', pool_recycle=3600)
    dbConnection    = sqlEngine.connect()
    try:
                       frame = dataFrame.to_sql(tableName, dbConnection, if_exists='replace');
    except ValueError as vx:
                       print(vx)
    except Exception as ex:
                       print(ex)
    else:
                       print("Table %s)created successfully."%dataframe);
    finally:
                       dbConnection.close()
```

```python
load_data()
```

```python
schedule.every(10).minutes.do(load_data)
```
